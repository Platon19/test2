```python
def poisk_min(matr, N, M):
    min=matr[0][0]
    for i in range(N):
        for j in range(M):
            if min>matr[i][j]:
                min=matr[i][j]
    return min
def poisk_max(matr, N, M):
    max=matr[0][0]
    for i in range(N):
        for j in range(M):
            if max<matr[i][j]:
                max=matr[i][j]
    return max
def obrez(matr, N, M):
    b=[]
    if N>M:
        for i in range(M):
            b.append(matr[i])
    if N<M:
        for i in range(N):
            b.append(matr[i][:N])
    return b
def transposition(matr, N, M):
    b=[]
    for i in range(M):
        b.append([])
    for i in range(M):
        k=0
        for j in range(N):
            b[i].append(matr[k][i])
            k+=1
    return b
def sum_zifr(a):
    b=0
    while a!=0:
        b=b+a%10
        a=a//10
    return b
def dopolnenie(matr, N, M):
    b=[]
    for i in range(M):
            b.append(0)
    while N>M:
        for i in range(N):
            matr[i].append(0)
        M+=1
    
    while N<M:
        matr.append(b)
        N+=1
    return matr


def clockwise(matr, N, M):
    matr=transposition(matr, N, M)
    x=0
    y=N-1
    while y-x>=1:
        matr=transposition(matr, N, M)
        matr[x], matr[y]=matr[y], matr[x]
        matr=transposition(matr, N, M)
        y=y-1
        x=x+1
    return matr

def counterclockwise(matr, N, M):
    matr=transposition(matr, N, M)
    x=0
    y=N-1
    while y-x>=1:
        matr[x], matr[y]=matr[y], matr[x]
        y=y-1
        x=x+1
    return matr


    
    
```
