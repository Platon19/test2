```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-10,10,1000)
y=np.arctan(x)
y1=np.arctan(2*x-4)
points=np.array([0, np.pi, 2*np.pi,  3*np.pi, 4*np.pi, 5*np.pi])
labels=[r'$0$',r'$\pi$',r'$2\pi$',r'$3\pi$',r'$4\pi$',r'$5\pi$']
plt.figure(figsize=(14,10))
plt.subplot(231)


area=np.linspace(0,4,1000)
y_area_top=np.arctan(area)
y_area_low=np.arctan(2*area-4)
plt.fill_between(area,y_area_top,y_area_low,color='orange')


plt.plot(x,y,color='red',lw=2, ls='solid',label=r'$y=\arctan(x)$')
plt.plot(x,y1,lw=2,label=r'$y=\arctan(2x-4)$',color='blue')

plt.xlim(0,5)
plt.grid()
plt.legend(loc=0)
plt.title('№1')
plt.xlabel('Oсь x')
plt.ylabel('Oсь y')

plt.subplot(232)


y=np.linspace(2,6,1000)
x=np.sqrt(6-y)
plt.fill_between(x,0,y,color='green')
plt.plot(x,y,color='blue',lw=2,label=r'$x=\sqrt{6-y}, y \geq 2$')
y1=np.linspace(0,2,1000)
x1=4-np.sqrt(2*y1)
plt.plot(x1,y1,color='red',lw=2,label=r'$x=4-\sqrt{2y}, y \leq 2$')
plt.fill_between(x1,0,y1,color='green')



plt.xlim(0,5)
plt.grid()
plt.title('№2')
plt.xlabel('Oсь x')
plt.ylabel('Oсь y')
plt.legend(loc=1)


plt.subplot(233,projection='polar')


fi=np.linspace(0,2*np.pi,1000)
r=1+np.cos(fi)
plt.polar(fi,r,color='red',label=r'$r=1-\cos(\varphi)$',lw=2)
r1=3*np.cos(fi)
plt.polar(fi,r1,color='blue',label=r'$r=3\cos(\varphi)$',lw=2)
fi_circle=np.linspace(np.pi/3,5*np.pi/3,1000)
r_circle=3*np.cos(fi_circle)
plt.fill_between(fi_circle,0,r_circle,color='orange')
fi_cardioid=np.linspace(-np.pi/3,np.pi/3,1000)
r_cardioid=1+np.cos(fi_cardioid)
plt.fill_between(fi_cardioid,0,r_cardioid,color='orange')



plt.title('№3')
plt.ylim(bottom=0)
plt.legend(loc=2)
ax = plt.gca()
ax.set_rgrids([i for i in range(4)])
ax.set_rlabel_position(210)


plt.subplot(234)


x=np.linspace(1,2,1000)
y=x*x/8-np.log(x)
plt.plot(x,y,lw=2,color='red',label=r'$y=\frac{x^2}{8}-\ln(x)$')


plt.legend()
plt.grid()
plt.title('№4')
plt.xlabel('Oсь x')
plt.ylabel('Oсь y')



plt.subplot(235)


x=np.linspace(-1,1,1000)
y=4+np.sqrt(1-np.square(x))
plt.plot(x,y,color='red',lw=2,label=r'$x^2+(y-4)^2=1$')
y1=4-np.sqrt(1-np.square(x))
plt.plot(x,y1,color='red',lw=2)
plt.fill_between(x,y1,y,color='yellow')


plt.legend(loc=3)
plt.grid()
plt.title('№5')
plt.xlabel('Oсь x')
plt.ylabel('Oсь y')
plt.xlim(-3,3)
plt.ylim(0,6)



plt.subplot(236)





x=np.linspace(10000,100000,1000)
y=(np.log(1+1/x)*np.sqrt(x))/(0.01*x+1)
plt.plot(x,y,color="red",lw=2,label=r'$y=\frac{\ln(1+\frac{1}{x})\sqrt{x}}{0.01x+1}$')
x1=np.linspace(10,100,1000)
y1=1/x
plt.plot(x,y1,color="blue",lw=2,label=r'$y=\frac{1}{x}}$')


plt.legend()
plt.grid()
plt.title('№6(а)')
plt.xlabel('Oсь x')
plt.ylabel('Oсь y')




plt.show()
```

```python
plt.figure(figsize=(14,10))

x=np.linspace(0.006,0.07,1000)
y=1/(np.sin(np.sqrt(x))*np.sqrt(np.sin(x)))
plt.plot(x,y,color="yellow",lw=9,label=r'$y=\frac{1}{\sqrt{\sin(x)}\sin(\sqrt{x})}$')
y1=1/x
plt.plot(x,y1,color="blue",lw=2,label=r'$y=\frac{1}{x}}$')


plt.legend()
plt.grid()
plt.title('№6(б)')
plt.xlabel('Oсь x')
plt.ylabel('Oсь y')
plt.xlim(0,0.02)
plt.ylim(-10,170)




plt.show()
```

```python
plt.figure(figsize=(14,10))

x=np.linspace(0.01,10,1000)
y=(1-np.cos(x*x+x))/np.sqrt(x**5)
plt.plot(x,y,color="red",lw=2,label=r'$y=\frac{1-\cos(x^2+x)}{\sqrt{x^5}}$')
x1=np.linspace(0.2,10,1000)
y1=1/x1
plt.plot(x1,y1,color="blue",lw=2,label=r'$y=\frac{1}{x}}$')


plt.legend()
plt.grid()
plt.title('№6(в)')
plt.xlabel('Oсь x')
plt.ylabel('Oсь y')





plt.show()
```
